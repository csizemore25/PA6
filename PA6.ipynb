```python
import math
```

```python
class calc(): 
    def __init__(self, input1, input2): 
        self.input1=input1
        self.input2=input2
        
    def addition(num1, num2): 
        print(f"Addition: {num1+num2}")
        
    def subtraction(num1, num2): 
        print(f"Subtraction: {num1-num2}")
        
    def multiplcation(num1, num2): 
        print(f"Multiplcation: {num1*num2}")
        
    def division(num1, num2): 
        print(f"Division: {num1/num2}")
    
    def squared(num1): 
        print(f"Squared: {num1**2}")
        
    def square_root(num1): 
        print(f"Square Root: {math.sqrt(num1)}")
        
    def power(num1, num2): 
        print(f"Power: {math.pow(num1, num2)}")
        
    def radians(num1): 
        print(f"Radians: {math.radians(num1)}")
        
    def degrees(num1): 
        print(f"Degrees: {math.degrees(num1)}")
        
    def sin(num1): 
        print(f"Sin: {math.sin(num1)}")
    
    def cos(num1): 
        print(f"Cos: {math.cos(num1)}")
        
    def tan(num1): 
        print(f"Tan: {math.tan(num1)}")
    
    def log(num1, num2): 
        print(f"Log: {math.log(num1, num2)}")
        
    def factorial(num1): 
        print(f"Factorial: {math.factorial(num1)}")
        
    def abs_value(num1): 
        print(f"Absolute Value: {math.fabs(num1)}")
        
        
while True: 
    print("Choose from the following menu: ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplcation")
    print("4. Division")
    print("5. Squared")
    print("6. Square Root")
    print("7. Power")
    print("8. Radians")
    print("9. Degrees")
    print("10. Sin")
    print("11. Cos")
    print("12. Tan")
    print("13. Log")
    print("14. Factorial")
    print("15. Absolute Value")
    
    choice=int(input("Enter choice (1-16): "))
    
    if choice==1: 
        num1=int(input("Enter first number: "))
        num2=int(input("Enter second number: "))
        calc.addition(num1,num2)
    elif choice==2: 
        num1=int(input("Enter first number: "))
        num2=int(input("Enter second number: "))
        calc.subtraction(num1,num2)
    elif choice==3: 
        num1=int(input("Enter first number: "))
        num2=int(input("Enter second number: "))
        calc.multiplcation(num1,num2)
    elif choice==4: 
        num1=int(input("Enter first number: "))
        num2=int(input("Enter second number: "))
        if num2==0: 
            print("Infinity")
        else: 
            calc.division(num1,num2)
    elif choice==5: 
        num1=int(input("Enter the value: "))
        calc.squared(num1)
    elif choice==6: 
        num1=int(input("Enter the value: "))
        calc.square_root(num1)
    elif choice==7: 
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        calc.power(num1, num2)
    elif choice==8:
        num1=int(input("Enter the angle you'd like to convert to radians: "))
        calc.radians(num1)
    elif choice==9: 
        num1=int(input("Enter the radian you'd like to convert to degrees: "))
        calc.degrees(num1)
    elif choice==10: 
        num1=int(input("Enter the value: "))
        calc.sin(num1)
    elif choice==11: 
        num1=int(input("Enter the value: "))
        calc.cos(num1)
    elif choice==12: 
        num1=int(input("Enter the value: "))
        calc.tan(num1)
    elif choice==13: 
        num1=int(input("Enter the value: "))
        num2=int(input("Enter the base of the log: "))
        if num1<=0 or num2<=0: 
            print("Invalid inputs") 
        else: 
            calc.log(num1, num2)
    elif choice==14: 
        num1=int(input("Enter the value: "))
        calc.factorial(num1)
    elif choice==15: 
        num1=int(input("Enter the value: "))
        calc.abs_value(num1)
    elif choice==16: 
        break
    else: 
        print("Wrong choice")
              
```
